```python
import geopandas as gpd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# open the segments

CiboLRUSegmentsAOI1 = gpd.read_file('C:/Users/Juan/Cibo Labs Pty Ltd/CiboSharepointArchive - CiboSharepointArchive/Share/LoveToGlobal/StratificationTest/CiboLRUSegmentsAOI1.geojson')
CiboLRUSegmentsAOI1
```

```python
# show me some statistics
CiboLRUSegmentsAOI1.describe()
```

```python
CiboLRUSegmentsAOI1.columns
```

```python
# plot the first 4 PCAs just to see if there are any obvious groups

sns.pairplot(data=CiboLRUSegmentsAOI1[['rasterAttributes_pca_Band_1_mean',
                                       'rasterAttributes_pca_Band_2_mean',
                                       'rasterAttributes_pca_Band_3_mean',
                                       'rasterAttributes_pca_Band_4_mean']],
             kind="hist",
             diag_kind='hist',
             markers='.'
             )
plt.show()
```

```python
# now cluster into 5 classes using k-means. Input variables are the 11 PCAs and their stdev 

data_for_clustering = CiboLRUSegmentsAOI1[['rasterAttributes_pca_Band_1_mean',
       'rasterAttributes_pca_Band_1_std', 'rasterAttributes_pca_Band_2_mean',
       'rasterAttributes_pca_Band_2_std', 'rasterAttributes_pca_Band_3_mean',
       'rasterAttributes_pca_Band_3_std', 'rasterAttributes_pca_Band_4_mean',
       'rasterAttributes_pca_Band_4_std', 'rasterAttributes_pca_Band_5_mean',
       'rasterAttributes_pca_Band_5_std', 'rasterAttributes_pca_Band_6_mean',
       'rasterAttributes_pca_Band_6_std', 'rasterAttributes_pca_Band_7_mean',
       'rasterAttributes_pca_Band_7_std', 'rasterAttributes_pca_Band_8_mean',
       'rasterAttributes_pca_Band_8_std', 'rasterAttributes_pca_Band_9_mean',
       'rasterAttributes_pca_Band_9_std', 'rasterAttributes_pca_Band_10_mean',
       'rasterAttributes_pca_Band_10_std', 'rasterAttributes_pca_Band_11_mean',
       'rasterAttributes_pca_Band_11_std']]

clusterer = KMeans(n_clusters=5, random_state=10)
cluster_labels = clusterer.fit_predict(data_for_clustering)
```

```python
# add the result to the input data so I can plot again
CiboLRUSegmentsAOI1['clusters'] = cluster_labels.astype(str)
```

```python
# plot the first 4 PCAs just to see if there are any obvious groups

sns.pairplot(data=CiboLRUSegmentsAOI1[['rasterAttributes_pca_Band_1_mean',
                                       'rasterAttributes_pca_Band_2_mean',
                                       'rasterAttributes_pca_Band_3_mean',
                                       'rasterAttributes_pca_Band_4_mean',
                                       'clusters']],
             hue='clusters',
             diag_kind='hist',
             markers='.'
             )
plt.show()
```
